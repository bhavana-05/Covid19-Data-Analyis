## Data Visualization and Analysis

### 1. Scatter Plot - Relationship between data

#### a. New Cases vs Deaths on 11th May, 2021

```python
plt.figure(figsize=(6, 4))
plt.scatter(df_final['New_Cases_11May'], df_final['Deaths_11May'], c='red', alpha=0.7,  s=30, edgecolors='black')
plt.xlabel('New Cases on 11 May 2021')
plt.ylabel('Deaths on 11 May 2021')
plt.title('New Cases vs Deaths')
plt.grid(True, alpha=0.5)
#plt.tight_layout()
plt.savefig('scatter_new_cases_vs_deaths.png')
plt.show()
```

#### b. Active Cases vs New Cases

```python
plt.figure(figsize=(6, 4))
plt.scatter(df_final['Active_Cases_Till_10May'], df_final['New_Cases_11May'], c='red', alpha=0.7,  s=30, edgecolors='black')
plt.xlabel('Active Cases till 10 May 2021')
plt.ylabel('New Cases on 11 May 2021')
plt.title('Active cases vs New Cases')
plt.grid(True, alpha=0.5)
plt.savefig('scatter_active_cases_vs_new_cases.png')
plt.show()
```

#### c. New Cases Vs Discharged cases

```python
plt.figure(figsize=(6, 4))
plt.scatter(df_final['New_Cases_11May'], df_final['Discharged_11May'], c='red', alpha=0.7,  s=30, edgecolors='black')
plt.ylabel('Discharged on 11 May 2021')
plt.xlabel('New Cases on 11 May 2021')
plt.title('New Cases vs Discharged')
plt.grid(True, alpha=0.5)
plt.savefig('scatter_new_cases_vs_discharged.png')
plt.show()
```

#### Satter plot analysis

```python
# Print correlation coefficient
corr_newcases_deaths = df_final['New_Cases_11May'].corr(df_final['Deaths_11May'])
print(f"Correlation coefficient between New Cases and Deaths: {corr_newcases_deaths:.3f}")

corr_active_newcases = df_final['Active_Cases_Till_10May'].corr(df_final['New_Cases_11May'])
print(f"Correlation coefficient between Active Cases and New Cases: {corr_active_newcases:.3f}")

corr_newcases_discharged = df_final['New_Cases_11May'].corr(df_final['Discharged_11May'])
print(f"Correlation coefficient between New Cases and Discharged: {corr_newcases_discharged:.3f}")
```

### 2. Box plots - Understanding Distributions

#### a. New Cases on 11 may

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df_final['New_Cases_11May'])
plt.xlabel('New Cases on 11 May 2021')
plt.title('Distribution of New Cases')
#plt.grid(True, alpha=0.5)
plt.savefig('boxplot_new_cases.png')
plt.show()
```

#### b. Deaths on 11May

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df_final['Deaths_11May'])
plt.xlabel('Deaths on 11 May 2021')
plt.title('Distribution of Deaths')
#plt.grid(True, alpha=0.5)
plt.savefig('boxplot_deaths.png')
plt.show()
```

#### c. Comparision of New total active and old active

```python
# Multiple box plots (comparing columns)
plt.figure(figsize=(8, 6))
plt.boxplot([df_final['Active_Cases_Till_10May'], df_final['Active_Cases_11May']], 
            tick_labels=['Active on 10May', 'Active on 11May'],
            patch_artist=True)
plt.ylabel('Count')
plt.title('Comparision of New total active cases and old active cases')
plt.savefig('boxplot_active_cases_comparison.png')
plt.show()
```

#### Box plot analysis

```python
# Outliers for new cases, using IQR - which cities have very high new cases
print("Outlier Detection for New Cases on 11 May 2021:")
Q1 = df_final['New_Cases_11May'].quantile(0.25)
Q3 = df_final['New_Cases_11May'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier threshold
outlier_threshold = Q3 + 1.5 * IQR

# Identify outliers
outliers = df_final[df_final['New_Cases_11May'] > outlier_threshold]
print(outliers[['District', 'New_Cases_11May']])

#Outliers for deaths, using IQR - which cities have very high new deaths
print("Outlier Detection for Deaths on 11 May 2021:")
Q1_deaths = df_final['Deaths_11May'].quantile(0.25)
Q3_deaths = df_final['Deaths_11May'].quantile(0.75)
IQR_deaths = Q3_deaths - Q1_deaths

# Define outlier threshold
outlier_threshold_deaths = Q3_deaths + 1.5 * IQR_deaths

# Identify outliers
outliers_deaths = df_final[df_final['Deaths_11May'] > outlier_threshold_deaths]
print(outliers_deaths[['District', 'Deaths_11May']])
```

### 3. Line Plots - District wise trends

#### a. District wise New cases

```python
# Sort districts
df_sorted = df.sort_values('New_Cases_11May', ascending=False)

plt.figure(figsize=(15, 8))
plt.plot(range(len(df)), df_sorted['New_Cases_11May'], 
         marker='o', linewidth=2, markersize=5, color='orange')

plt.xlabel('Districts')
plt.ylabel('New Cases on 11 May 2021')
plt.title('District-wise New Cases on 11 May 2021')

plt.xticks(range(len(df_sorted)), df_sorted['District'], 
           rotation=45, ha='right')

plt.grid(True, alpha=0.3)
plt.tight_layout()  
plt.savefig('lineplot_new_cases_districtwise.png')
plt.show()
```

```python
# Sort districts by new cases and get top 15
# one row is total count, so take 16
df_top15 = df_sorted.head(16)

plt.figure(figsize=(12, 6))
plt.plot(range(len(df_top15)), df_top15['New_Cases_11May'], 
         marker='o', linewidth=2, markersize=5, color='orange')

plt.xlabel('Districts')
plt.ylabel('New Cases on 11 May 2021')
plt.title('Top 15 Districts by New Cases on 11 May 2021')


plt.xticks(range(len(df_top15)), df_top15['District'], 
           rotation=45, ha='right')

plt.grid(True, alpha=0.3)
plt.tight_layout() 
plt.savefig('lineplot_top15_new_cases_districtwise.png')
plt.show()
```

#### b. Top 15 District wise Deaths

```python
# Sort districts by deaths
df_top15_deaths = df.sort_values('Deaths_11May', ascending=False).head(16)

plt.figure(figsize=(12, 6))
plt.plot(range(len(df_top15_deaths)), df_top15_deaths['Deaths_11May'], 
         marker='o', linewidth=2, markersize=6, color='red')

plt.xlabel('Districts')
plt.ylabel('Deaths on 11 May 2021')
plt.title('Top 15 Districts by Deaths on 11 May 2021')

# Set district names as x-tick labels with rotation
plt.xticks(range(len(df_top15_deaths)), df_top15_deaths['District'], 
           rotation=45, ha='right')

plt.grid(True, alpha=0.3)
plt.tight_layout()  
plt.savefig('lineplot_top15_deaths_districtwise.png')
plt.show()
```

#### c. Comparison of Active Cases: 10th May vs 11th May

```python
# Get top 20 districts by active cases on 11th May
df_top20_active = df.sort_values('Active_Cases_11May', ascending=False).head(21)

plt.figure(figsize=(14, 8))

# Plot both lines on the same plot
plt.plot(range(len(df_top20_active)), df_top20_active['Active_Cases_Till_10May'], 
         marker='o', linewidth=1.5, markersize=5, color='blue', label='Active Cases (10 May)')

plt.plot(range(len(df_top20_active)), df_top20_active['Active_Cases_11May'], 
         marker='o', linewidth=1.5, markersize=5, color='red', label='Active Cases (11 May)')

plt.xlabel('Districts')
plt.ylabel('Active Cases')
plt.title('Comparison of Active Cases: 10th May vs 11th May (Top 20 Districts)')

# Set district names as x-tick labels with rotation
plt.xticks(range(len(df_top20_active)), df_top20_active['District'], 
           rotation=45, ha='right')

# Add legend to distinguish the two lines
plt.legend()

plt.grid(True, alpha=0.3)
plt.tight_layout()  
plt.savefig('lineplot_top20_active_cases_comparison.png')
plt.show()

```
